```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos desde el CSV generado en "21Adquisition.ipynb"
merged_data = pd.read_csv('merged_data1.csv', low_memory=False)
```

```python
# Visualización de la distribución de las ventas
plt.figure(figsize=(10, 6))
sns.histplot(merged_data['Sales'], bins=50, kde=True)
plt.title('Distribución de las Ventas')
plt.xlabel('Ventas')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Visualización de la relación entre ventas y clientes
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Customers', y='Sales', data=merged_data)
plt.title('Relación entre Ventas y Clientes')
plt.xlabel('Clientes')
plt.ylabel('Ventas')
plt.show()
```

```python
# Visualización de la relación entre ventas y promociones
plt.figure(figsize=(10, 6))
sns.boxplot(x='Promo', y='Sales', data=merged_data)
plt.title('Relación entre Ventas y Promociones')
plt.xlabel('Promoción')
plt.ylabel('Ventas')
plt.show()
```

```python
# Visualización de la relación entre ventas y el tipo de tienda
plt.figure(figsize=(10, 6))
sns.boxplot(x='StoreType', y='Sales', data=merged_data)
plt.title('Relación entre Ventas y Tipo de Tienda')
plt.xlabel('Tipo de Tienda')
plt.ylabel('Ventas')
plt.show()
```

```python
#Ventas por día de la semana, siendo 1 el lunes y el 7 el domingo
sales_by_dow = merged_data.groupby('DayOfWeek')['Sales'].sum().reset_index()
sns.barplot(x='DayOfWeek', y='Sales', data=sales_by_dow)
plt.title('Ventas por Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Ventas')
plt.show()
```

```python
#Analisis de correlacion entre las variables numericas
numeric_data = merged_data.select_dtypes(include=[np.number])
correlation_matrix = numeric_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

```python
# Analisis de fechas, ventas a lo largo del tiempo
merged_data['Date'] = pd.to_datetime(merged_data['Date'])
sales_over_time = merged_data.groupby('Date')['Sales'].sum().reset_index()
plt.figure(figsize=(12, 6))
plt.plot(sales_over_time['Date'], sales_over_time['Sales'])
plt.title('Ventas a lo Largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.xticks(rotation=45)
plt.show()
```

```python
#Análisis de variables categóricas: ventas por tipo de tienda con un grafico de barras
sales_by_store_type = merged_data.groupby('StoreType')['Sales'].sum().reset_index()
sns.barplot(x='StoreType', y='Sales', data=sales_by_store_type)
plt.title('Ventas por Tipo de Tienda')
plt.xlabel('Tipo de Tienda')
plt.ylabel('Ventas')
plt.show()
```
